```python
#04/07/2026
import pandas as pd
print("Q1:")
data = {
    'Name': ['Amit', 'Priya', 'Rohan', 'Sneha', 'Vikram'],
    'Age': [19, 21, 20, 22, 19],
    'City': ['Mumbai', 'Pune', 'Mumbai', 'Delhi', 'Pune'],
    'Marks': [85, 78, 92, 65, 88]
}
df = pd.DataFrame(data)
```

```python

print(" 2. Complete DataFrame:")
print(df)
```

```python
print(" 3. First 3 Rows:")
print(df.head(3))
```

```python
print("4. Last 2 Rows:")
print(df.tail(2))
```

```python
print("5. Shape of DataFrame :")
print(df.shape)
```

```python
print("6. Column Names :")
print(df.columns.tolist())
```

```python
print("7. Dataset Information:")
df.info()
```

```python
print("8. Statistical Summary:")
print(df.describe())
```

```python
print("11. Name Column Only:")
print(df['Name'])
```

```python
print(" 12. Name and Marks Columns:")
print(df[['Name', 'Marks']])
```

```python
print("13. Marks > 80 :")
print(df[df['Marks'] > 80])
```

```python
print(" 14. Age > 20:")
print(df[df['Age'] > 20])
```

```python
print("15. Students from Pune:")
print(df[df['City'] == 'Pune'])
```

```python
print("16. Marks between 70 and 90 :")
print(df[df['Marks'].between(70, 90)])
```

```python
print("17. City is Mumbai and Marks > 75:")
print(df[(df['City'] == 'Mumbai') & (df['Marks'] > 75)])
```

```python
print(" 18. Sorted by Marks (Ascending):")
print(df.sort_values(by='Marks', ascending=True))
```

```python
print(" 19. Sorted by Marks (Descending) :")
print(df.sort_values(by='Marks', ascending=False))
```

```python
print("20. Unique Cities:")
print(df['City'].unique())
```

```python
print("21. Student Count per City:")
print(df['City'].value_counts())
```

```python
print("22. Duplicate Records:")
print(df[df.duplicated()])
```

```python
print("23. Removing Duplicates :")
df_no_duplicates = df.drop_duplicates()
print(df_no_duplicates)
```

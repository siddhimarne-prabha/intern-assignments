```python
"""Question 6
Compare the marks of Student A and Student B on the same graph.
Student A:
[75, 80, 78, 90, 88]
Student B:
[70, 85, 80, 86, 92]
Requirements:
Different Colors
Legend
Grid
"""

import matplotlib.pyplot as plt
student_a = [75, 80, 78, 90, 88]
student_b = [70, 85, 80, 86, 92]

plt.plot(student_a, color='blue', label='Student A')
plt.plot(student_b, color='orange', label='Student B')
plt.legend()
plt.grid(True)
plt.show()
```

```python
"""Question 9
Create a line chart for website visitors over seven days.
Visitors:
[150, 180, 170, 200, 230, 210, 250]
Requirements:
Figure Size (10,5)
Black Line
Plus (+) Marker
Title
Legend
Grid
"""
import matplotlib.pyplot as plt
visitors = [150, 180, 170, 200, 230, 210, 250]

plt.figure(figsize=(10, 5))
plt.plot(visitors, color='black', marker='+', label='Daily Visitors')
plt.title("Website Visitors Over 7 Days")
plt.legend()
plt.grid(True)
plt.show()
```

```python
"""Question 10
Create a professional Monthly Sales Report.
Months:
["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
Sales:
[25000, 30000, 28000, 40000, 45000, 50000]
Requirements:
Figure Size (10,5)
Green Line
Circle (o) Marker
Dashed (--) Line Style
Line Width = 3
Chart Title
X-axis Label
Y-axis Label
Grid
Legend

"""
import matplotlib.pyplot as plt
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
sales= [25000, 30000, 28000, 40000, 45000, 50000]

plt.figure(figsize=(10, 5))
plt.plot(months, sales,
    color='green',
    marker='o',
    linestyle='--',
    linewidth=3,
    label='Sales'
)
plt.title("Monthly Sales Report")
plt.xlabel("Months")
plt.ylabel("Sales (in ₹)")
plt.grid(True)
plt.legend()
plt.show()
```

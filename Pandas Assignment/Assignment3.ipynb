```python

#24/06/2026
import pandas as pd
import numpy as np


data = {
    "Student ID": [1, 2, 3, 4, 2, 5],
    "Name": ["Ananya", "Rohan", "Siddharth", "Isha", "Rohan", "Kabir"],
    "Age": [15, np.nan, 14, 16, np.nan, 15],
    "Marks": [85, 90, np.nan, 92, 90, np.nan]
}

df = pd.DataFrame(data)
print("Original DataFrame")
print(df)
```

```python
# 1. Find missing values
print("1. Find Missing values:")
print(df.isnull())
```

```python
# 2. Count missing values per column
print("2. Count of Missing Values in every  Column:")
print(df.isnull().sum())
```

```python
# 3. Replace missing values with 0
print(" 3. DataFrame with Missing Values Replaced by 0:")
df_zeros = df.fillna(0)
print(df_zeros)
```

```python
# 4. Replace missing age with average age
print(" 4. Replacing Missing Age with Average Age\n ")
# Calculate the mean of the Age column
mean_age = df["Age"].mean()
print(f"Calculated Mean Age:", mean_age)
# Fill missing values only in the Age column
df["Age"] = df["Age"].fillna(mean_age)
print(df)
```

```python
print("5. Identify Duplicate Records\n")
print(df.duplicated())
```

```python
print(" 6. DataFrame after Removing Duplicate Records\n")
df_cleaned = df.drop_duplicates(keep="first")
print(df_cleaned)
```

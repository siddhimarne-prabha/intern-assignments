```python
#08/07/2026
import numpy as np
arr_1d = np.arange(1, 21)
print(f" 1D Array:\n{arr_1d}")
```

```python
print(" Q1: Display the first element:")
print("First element:", arr_1d[0])
```

```python
print("Q2: Display the last element using negative indexing:")
print("Last element:", arr_1d[-1])
```

```python
print("Q3: Display the third element of an array")
print("Third element (index 2):", arr_1d[2])
```

```python
print("Q4: Display the last five elements using slicing:")
print("Last five elements:", arr_1d[-5:])
```

```python
print(" Q5: Display elements from index 3 to 8:")
print("Elements from index 3 to 8:", arr_1d[3:8])
```

```python
matrix_4x4 = np.arange(1, 17).reshape(4, 4)
print(f"Base 4x4 Matrix:\n{matrix_4x4}")
```

```python
print("Q6: Display the first row:")
print(matrix_4x4[0, :])
```

```python
print("Q7: Display the last row of the matrix:")
print(matrix_4x4[-1, :])
```

```python
print(" Q8: Display the first column:")
print(matrix_4x4[:, 0])
```

```python
print("Q9: Display the last column :")
print(matrix_4x4[:, -1])
```

```python
print("Q10: Display the middle two rows:")
print(matrix_4x4[1:3, :])
```

```python
print("Q11: Display the middle two columns :")
print(matrix_4x4[:, 1:3])
```

```python
print("Q12: Display the first 2 x 2 sub-matrix:")
print(matrix_4x4[0:2, 0:2])
```

```python
print("Q13: Display the last 2 x 2 sub-matrix:")
print(matrix_4x4[2:4, 2:4])
```

```python
arr_3d = np.arange(1, 19).reshape(2, 3, 3)
print(f" 3D Array:\n{arr_3d}")
```

```python
print(" Q14: Create a 3D array and display the first matrix:")
print(arr_3d[0, :, :])
```

```python
print("Q15: Display the second matrix of a 3D array :")
print(arr_3d[1, :, :])
```

```python
print("Q16: Display a specific element from a 3D array :")
# Accesses the 2nd matrix, 3rd row, 2nd column (number 17)
print("Element at index [1, 2, 1]:", arr_3d[1, 2, 1])
```

```python
arr_bool = np.arange(1, 21)
print(f"Base Array for Boolean Indexing:\n{arr_bool}")
```

```python
print("Q17: Display only even numbers using Boolean Indexing:")
print(arr_bool[arr_bool % 2 == 0])
```

```python
print("Q18: Display only odd numbers using Boolean Indexing:")
print(arr_bool[arr_bool % 2 != 0])
```

```python
print("Q19: Display numbers greater than 10 using Boolean Indexing:")
print(arr_bool[arr_bool > 10])
```

```python
print(" Q20: Display numbers less than 15 using Boolean Indexing :")
print(arr_bool[arr_bool < 15])
```

```python
print("Q21: Use Fancy Indexing to display elements at index positions 0, 2, 4, 6:")
indices_1 = [0, 2, 4, 6]
print(arr_1d[indices_1])
```

```python
print("Q22: Use Fancy Indexing to display the first and last elements:")
indices_2 = [0, -1]
print(arr_1d[indices_2])
```

```python
print("Q23: Create a 5 x 5 matrix:")
matrix_5x5 = np.arange(1, 26).reshape(5, 5)
print(f" 5x5 Matrix:\n{matrix_5x5}")
print("First row:", matrix_5x5[0, :])
print("Last row:", matrix_5x5[-1, :])
print("First column:", matrix_5x5[:, 0])
print("Last column:", matrix_5x5[:, -1])
```

```python
print("Q24: Create a 4 x 5 matrix and display only the first 3 columns:")
matrix_4x5 = np.arange(1, 21).reshape(4, 5)
print(f"Base 4x5 Matrix:\n{matrix_4x5}")
print("First 3 columns:\n", matrix_4x5[:, 0:3])
```

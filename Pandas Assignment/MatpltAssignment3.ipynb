```python
#23/07/2026
#Part A – subplot() (Questions 1–6)
"""Q1.
Create a figure containing 2 subplots:
Line Chart
Bar Chart
"""
import matplotlib.pyplot as plt
plt.figure(figsize=(15, 10))
months = ['Jan', 'Feb', 'Mar', 'Apr']
sales = [10, 15, 7, 12]
plt.subplot(2,2,1)
plt.plot(months,sales, marker='o', color='blue')
plt.title("line chart")
plt.xlabel('Month')
plt.ylabel('Sales')
plt.grid(True)

plt.subplot(2,2,2)
plt.bar(months,sales, color='orange')
plt.title("bar chart")
plt.xlabel('Month')
plt.ylabel('Sales')
plt.grid(True, axis='y')
```

```python
"""Q2.
Create a figure containing 4 subplots:
Line Chart
Bar Chart
Pie Chart
Scatter Plot
"""
import matplotlib.pyplot as plt

# 1. Line Chart
plt.subplot(2, 2, 1)
plt.plot([1, 2, 3], [4, 5, 6])
plt.title('Line Chart')
plt.grid(True)

# 2. Bar Chart
plt.subplot(2, 2, 2)
plt.bar(['A', 'B', 'C'], [3, 7, 5])
plt.title('Bar Chart')
plt.grid(True, axis='y')

# 3. Pie Chart
plt.subplot(2, 2, 3)
plt.pie([30, 40, 30], labels=['X', 'Y', 'Z'], autopct='%1.1f%%')
plt.title('Pie Chart')

# 4. Scatter Plot
plt.subplot(2, 2, 4)
plt.scatter([1, 2, 3, 4], [10, 20, 15, 25])
plt.title('Scatter Plot')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
"""Q3.
Create a 2 × 2 Student Dashboard showing:
Student Marks
Attendance
Subject Distribution
Study Hours vs Marks
"""
import matplotlib.pyplot as plt

# 1. Student Marks
plt.subplot(2, 2, 1)
plt.plot(['Exam 1', 'Exam 2', 'Exam 3'], [85, 90, 88], marker='o', color='green')
plt.title('Student Marks')
plt.ylabel('Marks')
plt.grid(True)

# 2. Attendance
plt.subplot(2, 2, 2)
plt.bar(['Jan', 'Feb', 'Mar'], [95, 88, 92], color='blue')
plt.title('Attendance (%)')
plt.grid(True, axis='y')

# 3. Subject Distribution
plt.subplot(2, 2, 3)
plt.pie([30, 25, 25, 20], labels=['Math', 'Science', 'English', 'History'], autopct='%1.0f%%')
plt.title('Subject Distribution')

# 4. Study Hours vs Marks
plt.subplot(2, 2, 4)
plt.scatter([2, 4, 5, 7, 8], [55, 65, 75, 85, 95], color='purple')
plt.title('Study Hours vs Marks')
plt.xlabel('Hours')
plt.ylabel('Marks')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
"""Q4.
Create an Employee Dashboard containing:
Salary Trend
Department-wise Employees
Salary Distribution (Histogram)
Experience vs Salary
"""
import matplotlib.pyplot as plt

# 1. Salary Trend
plt.subplot(2, 2, 1)
plt.plot([2020, 2021, 2022, 2023], [50, 55, 63, 70], marker='s', color='navy')
plt.title('Salary Trend')
plt.grid(True)

# 2. Department-wise Employees
plt.subplot(2, 2, 2)
plt.bar(['HR', 'IT', 'Sales'], [10, 45, 25], color='coral')
plt.title('Department Employees')
plt.grid(True, axis='y')

# 3. Salary Distribution
plt.subplot(2, 2, 3)
plt.hist([45, 50, 52, 55, 60, 62, 65, 70, 72, 80, 85], bins=5, color='skyblue', edgecolor='black')
plt.title('Salary Distribution')
plt.grid(True)

# 4. Experience vs Salary
plt.subplot(2, 2, 4)
plt.scatter([1, 2, 3, 5, 7, 10], [40, 48, 52, 68, 85, 110], color='red')
plt.title('Experience vs Salary')
plt.xlabel('Years')
plt.ylabel('Salary ($k)')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
"""Q5.
Create a dashboard with:
Monthly Sales (Line)
Product Sales (Bar)
Payment Methods (Pie)
Customer Ratings (Histogram)
Use subplot() to display all charts together.
"""
import matplotlib.pyplot as plt

# 1. Monthly Sales
plt.subplot(2, 2, 1)
plt.plot(['Jan', 'Feb', 'Mar', 'Apr'], [100, 120, 110, 140], color='blue')
plt.title('Monthly Sales')
plt.grid(True)

# 2. Product Sales
plt.subplot(2, 2, 2)
plt.bar(['Prod A', 'Prod B', 'Prod C'], [300, 450, 200], color='green')
plt.title('Product Sales')
plt.grid(True, axis='y')

# 3. Payment Methods
plt.subplot(2, 2, 3)
plt.pie([50, 30, 20], labels=['Card', 'Cash', 'UPI'], autopct='%1.1f%%')
plt.title('Payment Methods')

# 4. Customer Ratings
plt.subplot(2, 2, 4)
plt.hist([4, 5, 3, 5, 4, 2, 5, 4, 3, 5, 4, 4], bins=5, color='yellow', edgecolor='black')
plt.title('Customer Ratings')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
"""Q6.
Create a 3 × 2 subplot layout and display any six different charts learned in Matplotlib.
"""
import matplotlib.pyplot as plt

# 1. Line
plt.subplot(3, 2, 1)
plt.plot([1, 2, 3], [2, 4, 1])
plt.title('1. Line Chart')
plt.grid(True)

# 2. Bar
plt.subplot(3, 2, 2)
plt.bar(['A', 'B'], [5, 10])
plt.title('2. Bar Chart')
plt.grid(True, axis='y')

# 3. Pie
plt.subplot(3, 2, 3)
plt.pie([40, 60], labels=['Group 1', 'Group 2'])
plt.title('3. Pie Chart')

# 4. Scatter
plt.subplot(3, 2, 4)
plt.scatter([1, 3, 5], [2, 5, 3])
plt.title('4. Scatter Plot')
plt.grid(True)

# 5. Histogram
plt.subplot(3, 2, 5)
plt.hist([1, 2, 2, 3, 3, 3, 4, 5], bins=5)
plt.title('5. Histogram')
plt.grid(True)

# 6. Horizontal Bar
plt.subplot(3, 2, 6)
plt.barh(['X', 'Y'], [15, 25])
plt.title('6. Horizontal Bar Chart')
plt.grid(True, axis='x')

plt.tight_layout()
plt.show()
```

```python
"""Q16.
Create the same Sales Line Chart using:
default
ggplot
dark_background
Compare the appearance.
"""
import matplotlib.pyplot as plt

# 1. Default Style
plt.style.use('default')
plt.figure(figsize=(6, 4))
plt.plot([1, 2, 3, 4], [10, 20, 15, 25], marker='o')
plt.title('Sales Line Chart - Style: default')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.grid(True)
plt.show()

# 2. ggplot Style
plt.style.use('ggplot')
plt.figure(figsize=(6, 4))
plt.plot([1, 2, 3, 4], [10, 20, 15, 25], marker='o')
plt.title('Sales Line Chart - Style: ggplot')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.grid(True)
plt.show()

# 3. Dark Background Style
plt.style.use('dark_background')
plt.figure(figsize=(6, 4))
plt.plot([1, 2, 3, 4], [10, 20, 15, 25], marker='o')
plt.title('Sales Line Chart - Style: dark_background')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.grid(True)
plt.show()
plt.tight_layout()

plt.style.use('default')
```

```python
import pandas as pd

data = {
    "Name": ["Aarav", "Isha", "Kabir", "Meera", "Rohan", "Ananya"],
    "Department": ["IT", "HR", "IT", "Finance", "HR", "Finance"],
    "Salary": [65000, 45000, 55000, 70000, 48000, 72000]
}

df = pd.DataFrame(data)
print("Original DataFrame")
print(df)
```

```python
# 1. Sort salary in ascending order
print(" 1. Sorted by Salary (Ascending):")
df_asc = df.sort_values(by="Salary", ascending=True)
print(df_asc)
```

```python
# 2. Sort salary in descending order
print(" 2. Sorted by Salary (Descending):")
df_desc = df.sort_values(by="Salary", ascending=False)
print(df_desc)
```

```python
# 3. Display unique departments
print("3. Unique Departments:")
unique_depts = df["Department"].unique()
print(unique_depts)
```

```python
# 4. Count employees in each department
print("4. Employee Count per DepartmentL:")
dept_counts = df["Department"].value_counts()
print(dept_counts)
```

```python
# 5. Find average salary of each department
print(" 5. Average Salary per Department :")
avg_salary = df.groupby("Department")["Salary"].mean()
print(avg_salary)
```

```python
# 6. Find maximum salary of each department
print(" 6. Maximum Salary per Department -")
max_salary = df.groupby("Department")["Salary"].max()
print(max_salary)
```

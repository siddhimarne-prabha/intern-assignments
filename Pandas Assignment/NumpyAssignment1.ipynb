```python
import numpy as np
#Q1
print(" Q1: Student Marks Array :")
marks = np.array([78, 85, 90, 67, 88, 92, 75, 81, 69, 95])
print("Complete array:", marks)
print("Data type:", marks.dtype)
print("Total number of elements:", marks.size)
```

```python
print("Q2: Employee IDs:")
employee_ids = np.array([101, 102, 103, 104, 105, 106])
print("Array:", employee_ids)
print("Number of dimensions:", employee_ids.ndim)
print("Shape:", employee_ids.shape)
```

```python
print("Q3: College Students Data (2D Array):")
students_data = np.array([
    [78, 82, 90],
    [65, 70, 88],
    [91, 86, 79]
])
print("Complete array:\n", students_data)
print("Shape:", students_data.shape)
print("Size (total elements):", students_data.size)
print("Data type:", students_data.dtype)
```

```python
print(" Q4: Zero Matrix:")
zeros_matrix = np.zeros((5, 5))
print("5x5 Zero Matrix:\n", zeros_matrix)
```

```python
print("Q5: Ones Matrix:")
ones_matrix = np.ones((4, 6))
print("4x6 Ones Matrix:\n", ones_matrix)
```

```python
print(" Q 6: Identity Matrix:")
identity_matrix = np.eye(6)
print("6x6 Identity Matrix:\n", identity_matrix)
```

```python
print("Q7: Full Matrix:")
full_matrix = np.full((4, 4), 25)
print("4x4 Matrix filled with 25:\n", full_matrix)
```

```python
print("Q8: arange():")
array1 = np.arange(1, 11)
array2 = np.arange(10, 101, 10)
print("Array 1 (1 to 10):", array1)
print("Array 2 (10 to 100):", array2)
```

```python
print(" Q 9: linspace():")
array = np.linspace(0, 100, 8)
print("linspace():\n",array)
```

```python
print("Q10: Complete Array Properties:")
array = np.array([
    [10, 20, 30],
    [40, 50, 60],
    [70, 80, 90]
])
print("Array:\n",array)
print("Number of Dimensions (ndim):",array.ndim)
print("Shape (shape):",array.shape)
print("Size (size):",array.size)
print("Data Type (dtype):",array.dtype)
```

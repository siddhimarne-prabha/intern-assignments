```python
import numpy as np
# Q1 - Q4
arr1 = np.array([10, 20, 30, 40, 50])
arr2 = np.array([5, 10, 15, 20, 25])
```

```python
print("Q1: Addition:")
print("Addition Result:", arr1 + arr2)
```

```python
print("Q2: Subtraction:")
print("Subtraction Result:", arr1 - arr2)
```

```python
print(" Q3: Multiplication:")
print("Multiplication Result:", arr1 * arr2)
```

```python
print(" Q4: Division:")
print("Division Result:", arr1 / arr2)
```

```python
print("Q5: Floor Division, Modulus, and Power:")
arr_q5 = np.array([12, 24, 36, 48, 60])
print("Floor Division by 5:", arr_q5 // 5)
print("Modulus by 5:", arr_q5 % 5)
print("Power of 2:", arr_q5 ** 2)
```

```python
print(" Q6: Add 5000 Bonus:")
salaries = np.array([25000, 30000, 35000, 40000, 45000])
print("Updated Salaries:", salaries + 5000)
```

```python
print("Q7: Increase Marks by 5")
marks = np.array([65, 70, 75, 80, 85])
print("Updated Marks:", marks + 5)
```

```python
print("Q9: Conditional Display:")
arr_q9 = np.array([50, 60, 70, 80, 90])
print("Greater than or equal to 70:", arr_q9[arr_q9 >= 70])
print("Less than or equal to 60:", arr_q9[arr_q9 <= 60])
```

```python
print("Q10: Display Elements Equal to 20 :")
arr_q10 = np.array([10, 20, 30, 20, 40, 20])
print("Elements equal to 20:", arr_q10[arr_q10 == 20])
```

```python
print("Q11: Even and Odd Numbers:")
arr_q11 = np.array([12, 15, 18, 21, 24, 27, 30])
print("Even numbers:", arr_q11[arr_q11 % 2 == 0])
print("Odd numbers:", arr_q11[arr_q11 % 2 != 0])
```

```python
print("Q12: Total and Average Marks :")
marks_q12 = np.array([75, 82, 90, 68, 88])
print("Total Marks:", np.sum(marks_q12))
print("Average Marks:", np.mean(marks_q12))
```

```python
print("Q13: Median, Minimum, and Maximum:")
print("Median:", np.median(marks_q12))
print("Minimum:", np.amin(marks_q12))
print("Maximum:", np.amax(marks_q12))
print("\n")
```

```python
print(" Q14: Salary Metrics:")
salaries_q14 = np.array([25000, 30000, 35000, 45000, 55000])
print("Total Salary:", np.sum(salaries_q14))
print("Average Salary:", np.mean(salaries_q14))
print("Highest Salary:", np.amax(salaries_q14))
print("Lowest Salary:", np.amin(salaries_q14))
```

```python
print(" Q15: Variance and Standard Deviation:")
arr_q15 = np.array([20, 30, 40, 50, 60])
print("Variance:", np.var(arr_q15))
print("Standard Deviation:", np.std(arr_q15))
```

```python
print("Q16: Ascending and Descending Order:")
arr_q16 = np.array([45, 12, 78, 25, 60, 30])
print("Ascending Order:", np.sort(arr_q16))
print("Descending Order:", np.sort(arr_q16)[::-1])
```

```python
print("Q17: Salaries Ascending:")
salaries_q17 = np.array([35000, 25000, 60000, 45000, 30000])
print("Salaries in Ascending Order:", np.sort(salaries_q17))
```

```python
print(" Q18: Salaries Highest to Lowest :")
print("Salaries in Descending Order:", np.sort(salaries_q17)[::-1])
```

```python
print(" Q19: Find Index Positions of 30:")
arr_q19 = np.array([10, 20, 30, 40, 50, 30])
print("Indices where value is 30:", np.where(arr_q19 == 30)[0])
```

```python
print(" Q20: Insertion Position of 18 :")
arr_q20 = np.array([5, 10, 15, 20, 25, 30])
print("Insertion position for 18:", np.searchsorted(arr_q20, 18))
```

```python
print("Q21: Insertion Position of 350 :")
arr_q21 = np.array([100, 200, 300, 400, 500])
print("Insertion position for 350:", np.searchsorted(arr_q21, 350))
```

```python
print("Q22: Display Unique Values :")
arr_q22 = np.array([10, 20, 20, 30, 40, 40, 50])
print("Unique Values:", np.unique(arr_q22))
```

```python
print("Q23: Display Unique Marks :")
marks_q23 = np.array([70, 80, 70, 90, 80, 95])
print("Unique Marks:", np.unique(marks_q23))
```

```python
print("=== MIXED PRACTICE ===")
print("--- Q24: Student Marks Analysis ---")
student_marks = np.array([78, 65, 90, 55, 88, 72, 95])
print("Total Marks:", np.sum(student_marks))
print("Average Marks:", np.mean(student_marks))
print("Highest Marks:", np.amax(student_marks))
print("Lowest Marks:", np.amin(student_marks))
print("Sort in Ascending Order:", np.sort(student_marks))
print("Sort in Descending Order:", np.sort(student_marks)[::-1])
```

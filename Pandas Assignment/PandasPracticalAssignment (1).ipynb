```python
#06/07/2026
import pandas as pd
import numpy as np

data = {
    'Name': ['Amit', 'Priya', 'Rohan', 'Sneha', 'Vikram', 'Priya'],
    'Age': [19, np.nan, 20, 22, 19, np.nan],
    'Gender': ['Male', 'Female', 'Male', 'Female', 'Male', 'Female'],
    'City': ['Mumbai', 'Pune', 'Mumbai', 'Delhi', 'Pune', 'Pune'],
    'Course': ['BSc', 'BCom', 'BSc', 'BA', 'BTech', 'BCom'],
    'Attendance': [92, 85, 78, 95, 88, 85],
    'Marks': [85, np.nan, 92, 65, np.nan, np.nan],
    'Purchased': [1, 0, 1, 0, 1, 0]
}
df = pd.DataFrame(data)
```

```python
#  Q1: Replace all missing values in the dataset with 0
df_q1 = df.fillna(0)
print(df_q1)
```

```python
print(" Q2: Replace missing values in the Marks column using average marks :")
df_q2 = df.copy()
df_q2['Marks'] = df_q2['Marks'].fillna(df_q2['Marks'].mean())
print(df_q2)
```

```python
print(" Q3: Replace missing values in the Age column using average age :")
df_q3 = df.copy()
df_q3['Age'] = df_q3['Age'].fillna(df_q3['Age'].mean())
print(df_q3)
```

```python
print("Q4: Remove all rows that contain missing values:")
df_q4 = df.dropna()
print(df_q4)
```

```python
print(" Q5: Find all duplicate records in the dataset:")
duplicate_rows = df[df.duplicated()]
print(duplicate_rows)
```

```python
print("Q6: Remove duplicate records from the dataset:")
df_q6 = df.drop_duplicates()
print(df_q6)
```

```python
df_working = df.copy()
df_working['Marks'] = df_working['Marks'].fillna(df_working['Marks'].mean())
df_working['Age'] = df_working['Age'].fillna(df_working['Age'].mean())
```

```python
print(" Q7: Display students whose marks are greater than the average marks:")
avg_marks = df_working['Marks'].mean()
print(f"Calculated Average Marks: {avg_marks}")
print(df_working[df_working['Marks'] > avg_marks])
```

```python
print("Q8: Display students whose attendance is greater than 90% :")
print(df_working[df_working['Attendance'] > 90])
```

```python
print("Q9: Display students whose marks are between 70 and 90 :")
print(df_working[df_working['Marks'].between(70, 90)])
```

```python
print("Q10: Display female students from Pune:")
print(df_working[(df_working['Gender'] == 'Female') & (df_working['City'] == 'Pune')])
```

```python
print(" Q11: Display male students whose marks are greater than 80 :")
print(df_working[(df_working['Gender'] == 'Male') & (df_working['Marks'] > 80)])
```

```python
print("Q12: Find the maximum marks of the students:")
print("Max Marks:", df['Marks'].max())
```

```python
print(" Q13: Find the minimum marks of the students:")
print("Min Marks:", df['Marks'].min())
```

```python
print(" Q14: Calculate the average marks of all students:")
print("Average Marks:", df['Marks'].mean())
```

```python
print(" Q15: Calculate the total marks of all students:")
print("Total Marks:", df['Marks'].sum())
```

```python
print(" Q16: Display all unique values from the Course column:")
print(df['Course'].unique())
```

```python
print(" Q17: Count the number of students in each city using value_counts():")
print(df['City'].value_counts())
```

```python
print(" Q18: Count the number of students in each course:")
print(df['Course'].value_counts())
```

```python
print("Q19: Group the dataset by City and calculate the average marks:")
print(df_working.groupby('City')['Marks'].mean())
```

```python
print("Q20: Group the dataset by Course and calculate statistical metrics using agg():")
course_stats = df_working.groupby('Course')['Marks'].agg(
    Total_Marks='sum',
    Average_Marks='mean',
    Minimum_Marks='min',
    Maximum_Marks='max',
    Total_Students='count'
)
print(course_stats)
```

```python
print(" Q21: Find the correlation between all numerical columns:")
numerical_df = df_working.select_dtypes(include=[np.number])
print(numerical_df.corr())
```

```python
print("Q22: Perform Manual Encoding on the Gender column (Male -> 1, Female -> 0):")
df_manual_encoded = df_working.copy()
df_manual_encoded['Gender'] = df_manual_encoded['Gender'].map({'Male': 1, 'Female': 0})
print(df_manual_encoded)
```

```python
print(" Q23: Apply One-Hot Encoding on the City and Course columns ")
df_one_hot = pd.get_dummies(df_working, columns=['City', 'Course'], dtype=int)
print(df_one_hot)
```

```python
print(" Q24: Create Features (X) and Target (y) and display both:")
X = df_working[['Age', 'Attendance', 'Marks']]
y = df_working['Purchased']

print("Features (X):")
print(X)
print("\nTarget (y):")
print(y)
```

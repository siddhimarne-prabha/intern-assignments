```python
#Mini Project
emp_salary = np.array([[25000, 30000, 35000],
                       [40000, 45000, 50000]])

print("1. Original Matrix:")
print(emp_salary)

print("\n2. Reshaped into 3x2 :")
reshaped_salary = emp_salary.reshape(3, 2)
print(reshaped_salary)

print("\n3. Flattened Reshaped Matrix:")
flattened_salary = reshaped_salary.flatten()
print(flattened_salary)

print("\n4. Transpose of Original Matrix :")
print(np.transpose(emp_salary))

print("\n5. Increased Salary by ₹2000 (Broadcasting):")
updated_salary = emp_salary + 2000
print(updated_salary)

print("\n 6. Copy :")
salary_copy = emp_salary.copy()
salary_copy[0, 0] = 99999
print("Original Matrix (Unchanged):\n", emp_salary)
print("Copied Matrix (Modified):\n", salary_copy)

print("\n 7. View :")
salary_view = emp_salary.view()
salary_view[0, 0] = 11111
print("Original Matrix:\n", emp_salary)
print("View Matrix:\n", salary_view)
```

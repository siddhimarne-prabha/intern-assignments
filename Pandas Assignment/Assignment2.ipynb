```python
#23/06/2026

import pandas as pd

data = {
    "Employee ID": [101, 102, 103, 104, 105],
    "Name": ["Rahul", "Priya", "Amit", "Siddharth", "Neha"],
    "Department": ["IT", "HR", "Finance", "IT", "Marketing"],
    "Salary": [45000, 38000, 52000, 60000, 42000],
    "City": ["Pune", "Mumbai", "Pune", "Bangalore", "Pune"]
}

df = pd.DataFrame(data)
```

```python
# 1. Display first 3 rows
print("First 3 Rows")
print(df.head(3))
```

```python
# 2. Display last 2 rows
print("Last 2 Rows ")
print(df.tail(2))
```

```python
# 3. Display dataset shape
print("Dataset Shape (Rows, Columns) ")
print(df.shape)
```

```python
# 4. Display column names
print("Column Names")
print(df.columns)
```

```python
print("Employees with Salary > 40000")
print(df[df["Salary"] > 40000])
```

```python
print("Employees from Pune")
print(df[df["City"] == "Pune"])
```

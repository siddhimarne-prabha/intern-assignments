```python
#22/06/2026

import pandas as pd

# 1. Create the DataFrame
data = {
    "Name": ["siddhi", "riya", "ishwari", "sahil", "sonu"],
    "Age": [15, 16, 15, 17, 16],
    "City": ["New York", "Los Angeles", "Chicago", "Houston", "Miami"],
    "Marks": [88, 92, 79, 95, 91]
}

df = pd.DataFrame(data)
print(df)
```

```python
# 2. Display the DataFrame
print("Original DataFrame:")
print(df)
```

```python
# 3. Save the DataFrame into a CSV file named students.csv

df.to_csv("students.csv", index=False)
print("DataFrame successfully saved to 'students.csv'.\n")
```

```python
# 4. Read the CSV file again
loaded_df = pd.read_csv("students.csv")
```

```python

"""
Q25. Final Practical Challenge
Create a Sales Analysis Dashboard containing:
     Product Sales (Bar Chart)
      Payment Method (Pie Chart)
      Student Marks Distribution (Histogram)
      Study Hours vs Marks (Scatter Plot)
Requirements:
Use plt.subplot() to display all 4 charts in one figure.
Add a main title using plt.suptitle().
Add titles to each chart.
Add labels wherever applicable.
Add grid to Bar Chart, Histogram, and Scatter Plot.
Set figure size to (12, 8).
"""
import matplotlib.pyplot as plt

# Data
products = ['Laptop', 'Mobile', 'Tablet', 'Watch']
sales = [150, 300, 80, 120]

payment_methods = ['Cash', 'UPI', 'Card', 'Net Banking']
payment_counts = [15, 45, 30, 10]

student_marks = [45, 58, 62, 75, 80, 82, 85, 88, 90, 92, 95, 98]

study_hours = [1, 2, 3, 4, 5, 6, 7, 8, 9]
marks_scored = [35, 45, 50, 60, 70, 78, 85, 92, 98]

# Create Figure
plt.figure(figsize=(12, 8))
plt.suptitle('Sales & Performance Analysis Dashboard', fontsize=16)

# Subplot 1: Product Sales Bar Chart
plt.subplot(2, 2, 1)
plt.bar(products, sales, color='teal', edgecolor='black')
plt.title('Product Sales')
plt.xlabel('Product')
plt.ylabel('Sales')
plt.grid(True)

# Subplot 2: Payment Method Pie Chart
plt.subplot(2, 2, 2)
plt.pie(payment_counts, labels=payment_methods, autopct='%1.1f%%', startangle=90)
plt.title('Payment Method Breakdown')

# Subplot 3: Marks Distribution Histogram
plt.subplot(2, 2, 3)
plt.hist(student_marks, bins=5, color='orange', edgecolor='black')
plt.title('Student Marks Distribution')
plt.xlabel('Marks')
plt.ylabel('Frequency')
plt.grid(True)

# Subplot 4: Study Hours vs Marks Scatter Plot
plt.subplot(2, 2, 4)
plt.scatter(study_hours, marks_scored, color='purple', marker='o')
plt.title('Study Hours vs Marks')
plt.xlabel('Hours Studied')
plt.ylabel('Marks')
plt.grid(True)

plt.tight_layout()
plt.show()
```
